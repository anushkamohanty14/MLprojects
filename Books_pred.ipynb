```python

ratingscounts= data.groupby("Book-Title")["Book-Rating"].count()
ct=ratingscounts[ratingscounts>50].index
fdata=data[data["Book-Title"].isin(ct)]
fdata
```

### data preparation

```python
book_pivot = fdata.pivot_table(columns='User-ID', index='Book-Title', values="Book-Rating")
```

```python
book_pivot
```

```python

book_pivot.fillna(0, inplace=True)
```

## Model Building

```python
from sklearn.neighbors import NearestNeighbors
model = NearestNeighbors(algorithm='brute')
model.fit(book_pivot)
```

```python
distances, suggestions = model.kneighbors(book_pivot.loc['I Know Why the Caged Bird Sings', :].values.reshape(1, -1))
```

```python
for i in range(len(suggestions)):
  print(book_pivot.index[suggestions[i]])
```
