### stemming

```python
# Dowloading the required libraries
import nltk                     #stemming is the nlp technique to findthe rootword or stemword ##Natural_lanugage_toolkit
nltk.download("stopwords")
```

```python
X=df['genres']
```

```python
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
from sklearn.feature_extraction.text import CountVectorizer
```

```python
ps = PorterStemmer()
stop_words = set(stopwords.words('english'))

tag = []

# Assuming X is a column of strings (e.g., X = df['text_column'])
for text in X:
    if isinstance(text, list):
        text = ' '.join(text)  # convert list to space-separated string
    words = text.split()
    filtered = [ps.stem(word) for word in words if word.lower() not in stop_words]
    review = ' '.join(filtered)
    tag.append(review)


# Now tag has cleaned strings
vectorizer = CountVectorizer()
tag = tag[:4802]  # trim the extra junk at the end
X_cv = vectorizer.fit_transform(tag).toarray()
features_name = vectorizer.get_feature_names_out()
```

```python
X_df=pd.DataFrame(X_cv,columns=features_name)
X_df
```

## cosine similarity


```python
#Cosine similarity- Here we are introducing cosine similarity to compare the similarity between the
# records of the features.it wrks on the correlation cosine angle.
from sklearn.metrics.pairwise import cosine_similarity
similarity =cosine_similarity(X_cv)
similarity
```

```python
similarity_df=pd.DataFrame(similarity,columns=df['title'],index=df['title'])
similarity_df
```

```python
sd=similarity_df.iloc[:10,:10]
plt.figure(figsize=(20,20))
sns.heatmap(sd,  annot=True, cmap="coolwarm")
```

## Model function

```python
def recommend(movie):

  try:


    print('𝐍𝐨𝐰 𝐥𝐨𝐚𝐝𝐢𝐧𝐠. . .')
    print(" ")
    movie = process.extractOne(movie,data['title'])[0]
    movie_index = process.extractOne(movie,data['title'])[2]
    movie_per=  process.extractOne(movie,data['title'])[1]
    if(movie_per>50):

      print('Selected movie -',movie)
      distance = similarity[movie_index]
      print('-'*20,'Recommendations','-'*20)
      movie_list = sorted(list(enumerate(distance)),reverse=True,key=lambda x:x[1])[1:6]

      for i in movie_list:
        print(data.iloc[i[0]].title)

  except:

    print('𝙔𝙤𝙪𝙧 𝙨𝙚𝙖𝙧𝙘𝙝 - 𝙙𝙞𝙙 𝙣𝙤𝙩 𝙢𝙖𝙩𝙘𝙝 𝙖𝙣𝙮 𝙢𝙤𝙫𝙞𝙚.')
    print("► Make sure that all words are spelled correctly.")
    print("► Try different keywords.")
    print("► Try more general keywords.")
```
